We want to add the object type to the start of each image name.

[source](https://stackoverflow.com/questions/37467561/renaming-multiple-files-in-a-directory-using-python)

```python

objects = os.listdir(dataDir)
```

```python
# in case there's an error in renaming

for index, objectType in enumerate(objects):
    os.rename(os.path.join(dataDir, ''.join([str(index),'.jpg'])), os.path.join(dataDir, objectType))
```

```python
for objectNames in objects:
  # print(objectNames)
    os.rename(os.path.join(dataDir, objectNames), os.path.join(dataDir, ''.join([str(objectNames), '.jpg'])))
```

Converting all the filenames

```python
for objectNames in objects:
  files = os.listdir(os.path.join(dataDir, objectNames))
  for filename in files:
    os.rename(os.path.join(dataDir, objectNames, filename), os.path.join(dataDir, objectNames, ''.join([objectNames, '_', filename])))
```

Displaying the number of images in each of the objects.

```python
for objectName in objects:
  print(f'"{objectName}" has {len(os.listdir(os.path.join(dataDir, objectName)))} images')
```

```python
def globFn(objects: str):
    #stores the file names in a list
    imageFilenames = glob.glob(dataDir + '/' + objects + '/' + objects + '_*.jpg') # Where the asterisk symbol acts as a wildcard
    return imageFilenames
```

Displaying a single image from the glob function

```python
plt.imshow(plt.imread(globFn(objects[0])[0]))
```

Storing the filenames

```python
birdFiles = globFn('bird')
dragon_snakeFiles = globFn('dragon_snake')
flowerFiles = globFn('flower')
mixFiles = globFn('mix')
small_animalsFiles = globFn('small_animals')
fox_tigerFiles = globFn('fox_tiger')
```

Reading the filenames and storing the data

```python
bird = []
for filename in birdFiles:
  bird.append(plt.imread(filename))
```
